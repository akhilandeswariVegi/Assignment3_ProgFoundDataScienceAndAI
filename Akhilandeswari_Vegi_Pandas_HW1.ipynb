### Task 11
Create a new column titled `CS5000 Grade` in the score DataFrame to calcualte the letter grade for each student in course CS 5000 using the following criteria
+ `A: score >= 90`
+ `B: 80 <= score < 90`
+ `C: 70 <= score < 80`
+ `D: 60 <= score < 70`
+ `F: score < 60`

```python
def calculate_grade(x):
    if x >= 90:
        return 'A'
    elif 80 <= x < 90:
        return 'B'
    elif 70 <= x < 80:
        return 'C'
    elif 60 <= x < 70:
        return 'D'
    elif x < 60:
        return 'F'
    
# Apply the function to calculate the letter grade for CS 5000 scores
scores_df["CS5000 Grade"] = scores_df["CS 5000"].apply(lambda x:calculate_grade(x))
scores_df

```

### Task 13
Using three courses' scores (assuming each course has 3 credits), calculate each student's GPA for Spring 2023, respectively. The following criteria should be used in the calculation
+ `A: 4.0`
+ `B: 3.0`
+ `C: 2.0`
+ `D: 1.0`
+ `F: 0.0`

For example, if a student's grades are A, B and C. That student's GPA should be calculated as

`(3 * 4 + 3 * 3 + 3 * 2) / 9 = 3.0`

Add a new column titled `GPA` in the scores DataFrame to store the GPA for each student

```python
def gpa(x):
    if (x >= 90):
        return 4.0
    if (x >= 80) & (x < 90):
        return 3.0
    if (x >= 70) & (x < 80):
        return 2.0
    if (x >= 60) & (x < 70):
        return 1.0
    return 0.0

# apply() applies a function to each element of a DataFrame
scores_df['GPA'] = scores_df.apply(lambda x: (((gpa(x['CS 5000']) * 3) +
                                                   (gpa(x['CS 5110']) * 3) +
                                                   (gpa(x['CS 5600']) * 3)) / 9), axis = 1)
scores_df
```

### Task 14
Insert a new column title `Status` to provide a student academic standing based on the GPA
+ `Honor Roll: GPA >= 3.5`
+ `Normal: 3.0 <= GPA < 3.5`
+ `In Probation: GPA < 3.0`

```python
def find_status(gpa):
    if gpa >= 3.5:
        return 'Honor Roll'
    elif 3.0 <= gpa < 3.5:
        return 'Normal'
    else:
        return 'In Probation'

# Apply the function to determine the status based on GPA
scores_df['Status'] = scores_df['GPA'].apply(find_status)
scores_df
```
